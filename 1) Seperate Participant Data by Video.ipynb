```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

# Seperating the Participant Data by Video

The code below is designed to process participant data stored in CSV files.

```python
# loop through participant data
participants_directory = glob.glob("./data/Participant_Data/*.csv")
for participant_file in participants_directory:

    # grab the participant data
    participant = participant_file.split('\\')[-1].split(".")[0]
    print(f"Processing participant {participant}")
    participant_data = pd.read_csv(participant_file)

    # drop unnecessary columns
    participant_data = participant_data.iloc[:, 2:]
    participant_data.drop(columns=["Participant"], inplace=True)

    videos = participant_data["vidnum"].unique().tolist()  # list of vid #s

    # seperate participant data by vid nuber
    for video in videos:
        vid_data = participant_data.loc[participant_data["vidnum"] == video]
        vid_data = vid_data.drop(columns=["vidnum"])

        # save the data
        folder = f"Video_{video}"
        if not os.path.exists(f"./data/Video_Data/{folder}"):
            os.makedirs(f"./data/Video_Data/{folder}")
        print(f"Saving participant {participant} video {video} data to ./data/Video_Data/{folder}/{participant}_data.csv")
        vid_data.to_csv(f"./data/Video_Data/{folder}/{participant}_data.csv", index=False)
```
